## *Acts and Sections data*

```python
acts_sec = pd.read_csv("/kaggle/input/court-data/acts_sections/acts_sections.csv", low_memory=False)
acts_sec.head()
```

```python
acts_sec.info()
```

*cleaning and processing the data*

```python
acts_sec.drop(columns=["bailable_ipc", "criminal"], axis=1, inplace=True)
acts_sec.dropna(inplace=True)
```

*Importing the pre-processed data file into the notebook*

```python
cases_3yrs = pd.read_csv("/kaggle/input/data-preprocessing/3yrs_cases.csv")
cases_3yrs.head()
```

```python
cases_3yrs = pd.merge(cases_3yrs, acts_sec, how='left', on='ddl_case_id')
cases_3yrs.head()
```
